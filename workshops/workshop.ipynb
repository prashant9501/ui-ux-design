# WORKSHOP PROBLEM STATEMENTS



### General Guidelines

* You can **pick 3–4 AI features** (not everything) and justify the trade-offs in their UX Summary slide.
* **Prioritise 3-4 AI features**; argue in the UX-summary slide why others were deferred.
* Please label every mock-up with **“AI touchpoint” call-outs** to satisfy the rubric on explicating AI.
* Also try to demo **one micro-interaction** per project (e.g., swap suggestion animation, OCR field highlight, scan progress ring). These can be quick Figma Smart-Animate tricks.
* You are also encouraged to do **role-playing walkthroughs**: one student is the user, one the narrator.
* Please do not forget to label every mock-up with the AI touchpoint (“Powered by CV model”, “LLM rationale pop-over”) – this anchors grading to the AI-UX link mandated in the workshop brief.&#x20;



## 1. “SignBridge” — Real-Time Sign-Language Interpreter

### Problem statement

* **Who**: Deaf or hard-of-hearing people (primary) plus hearing counterparts (secondary) in everyday 1-to-1 interactions (shops, clinics, public offices).
* **Pain**: Certified interpreters are scarce and expensive; text chat is too slow; paper & pen feel exclusionary.
* **Impact**: Missed information, social isolation, lower service quality.
* **Scope boundaries**: One phone per party; casual conversations (≤10 min); American Sign Language to start. No group calls.

### AI-enabled features you may leverage

| Layer           | Candidate capability                                                                      | Note for UX                                                         |
| --------------- | ----------------------------------------------------------------------------------------- | ------------------------------------------------------------------- |
| Vision          | **Hand-shape & gesture recognition** (MediaPipe Hands + transformer classifier)           | Shows confidence overlay so user can slow down / retry              |
| Language        | **ASL gloss → English natural language** via seq-to-seq LLM fine-tuned on sign-text pairs | Explain reformulations (“*signed ‘FINISH YOU?’ → ‘Are you done?’*”) |
| Speech          | **Neural TTS + on-device ASR** for hearing side                                           | Toggle captions vs audio                                            |
| Personalisation | **Few-shot incremental learning** for user-defined signs (names, jargon)                  | “Teach a new sign” flow                                             |

### Key screens & states

1. **On-boarding / Accessibility set-up**

   * Choose signed language, text size, speech speed.
2. **Live Conversation**

   * Split view: left = camera with hand-joint overlay; right = rolling transcript bubbles.
   * Visual confidence meter (green/amber/red).
   * Mic button for the hearing user → speech bubbles appear and auto-sign avatar (optional).
3. **New-Sign Trainer (modal)**

   * Record 3 examples → label → save to personal dictionary.
4. **Conversation History**

   * Timestamped transcript with export to PDF / email.
   * Star to save key phrases.

## 2. “ClinicCompass” — 5-Minute Symptom Triage & Slot Booker

### Problem statement

* **Who**: Small outpatient clinics with ≤5 physicians; patients book by phone or walk-in.
* **Pain**: Receptionists perform subjective triage; low-risk colds clog queues while urgent cases wait.
* **Cost**: Overtime, physician idle time, patient dissatisfaction.
* **Scope boundaries**: Adult primary-care complaints only; not a diagnostic tool, just urgency + routing.

### AI-enabled features

| Stage      | Capability                                                                      | UX implication                               |
| ---------- | ------------------------------------------------------------------------------- | -------------------------------------------- |
| Intake     | **OCR of insurance card** (camera)                                              | Auto-fills demographics; edit inline         |
| Dialogue   | **Clinical-LLM symptom checker** that generates follow-up questions dynamically | Conversational UI; show “why we ask” tooltip |
| Risk       | **Bayesian risk stratification** (low / medium / high urgency)                  | Color-coded badge, disclaimers               |
| Scheduling | **Reinforcement model** predicting no-show probability & load balancing         | Suggests tele-visit vs in-person slots       |

### Key screens

1. **Consent & Privacy Notice** – mandatory scroll; disable “Next” until 5 s elapsed.
2. **Adaptive Chat Triage**

   * Rich-text bubbles, yes/no chips, photo-upload for rash.
   * Progress bar (“3 min left”).
3. **Triage Result & Recommendations**

   * Urgency badge, rationale peek (“Because chest pain + age > 50”).
   * CTA buttons: *Book earliest in-person*, *Video consult*, *Call nurse*.
4. **Slot Picker**

   * Calendar with clinic load heatmap; shows copay estimate.
5. **Physician Handoff Sheet** (internal)

   * Structured summary → injects into EHR.

## 3. “MatSelect” — Material-Recommendation Wizard for R\&D Labs

### Problem statement

* **Who**: Materials scientists/engineers exploring candidates for a new battery cathode or lightweight alloy.
* **Pain**: Searching scattered PDFs and databases, manually cross-referencing properties. Days lost.
* **Scope boundaries**: Discovery phase only; procurement & testing out of scope.

### AI-enabled features

| Function        | Capability                                                              | UX design hook                               |
| --------------- | ----------------------------------------------------------------------- | -------------------------------------------- |
| Semantic search | **Embedding search** across Materials Project, PubChem, internal docs   | Explain “why ranked #1” (similar glide-path) |
| Prediction      | **Graph neural net** that estimates missing properties (e.g., band-gap) | Icon marks predicted vs measured values      |
| Reasoning       | **LLM explainer** translating property trade-offs into plain English    | “Chat about this material” side panel        |
| Workflow        | **Active-learning recommender** suggests next query to narrow space     | Inline chips: *“Filter by cost?”*            |

### Key screens

1. **Property Criteria Builder**

   * Multi-select sliders (e.g., Conductivity > 10 S/cm, Temp ≤ 200 °C).
   * Toggle to include predicted values.
2. **Ranked Results Table**

   * Columns: Material, key props, confidence bar.
   * Column header hover → definition tooltip.
3. **Material Detail Drawer**

   * Spider/radar chart; provenance links to papers.
   * Button: *Add to Comparison*.
4. **Comparison Basket**

   * Side-by-side cards; export to CSV.
5. **Insight Chat**

   * Natural-language Q\&A (“Why is LiMnO₂ less stable above 180 °C?”).

## 4. “Shelfr” — Computer-Vision Shelf Auditor for Grocery Stores

### Problem statement

* **Who**: Store associates in medium-size grocery chains.
* **Pain**: Manual shelf checks every 2 hours miss mid-cycle stock-outs; managers rely on gut feel.
* **Business loss**: 4-8 % revenue hit from out-of-stock (OOS) items.
* **Scope boundaries**: Ambient indoor lighting, standard 4-ft gondola shelves; no freezer sections.

### AI-enabled features

| Stage    | Capability                                                          | UX cue                                    |
| -------- | ------------------------------------------------------------------- | ----------------------------------------- |
| Imaging  | **Edge-device CV**: object detection + pose estimation vs planogram | Draw colored boxes: green = OK, red = gap |
| Forecast | **Time-series model** predicts when shelf will run empty            | Countdown badge (e.g., “OOS in 45 min”)   |
| Dispatch | **Task prioritization** based on lost-sales estimate                | Sort task list by revenue impact          |
| Learning | **Few-shot user feedback loop** (“Not a gap”)                       | Thumbs-up/down on overlay                 |

### Key screens

1. **Capture & Detect**

   * Live camera with overlay boxes; auto-snap after focus lock.
   * Progress spinner (“Analysing…1.2 s”).
2. **Gap Confirmation Modal**

   * List of suspected SKUs with thumbnail; user confirms / rejects.
3. **Restock Task Queue**

   * Cards show item, location, qty needed, countdown. Swipe to mark done.
4. **Analytics Dashboard (manager)**

   * KPI tiles: OOS rate, compliance %, lost-sales \$.
   * Heatmap by aisle and hour.
5. **Settings / Planogram Upload**

   * Drag-and-drop PDF/CSV; preview parsed shelf layout.

## 5. “EcoCart” — Personal Carbon-Footprint Checkout

### Problem statement

* **Who**: Online shoppers who want to reduce climate impact without spending hours researching.
* **Pain**: Footprint data is buried in PDF LCA reports; “green” labels are vague; swaps aren’t obvious at checkout.
* **Scope**: Only covers scope-3 emissions of physical products sold by the same retailer; shipping method selection is in scope, delivery-route optimisation out of scope.

### AI-enabled features

| Layer          | Capability                                                                          | UX Hook                                           |
| -------------- | ----------------------------------------------------------------------------------- | ------------------------------------------------- |
| Mapping        | **SKU→LCA matcher** using text embeddings + metadata rules                          | Shows CO₂e figure with uncertainty bar            |
| Estimation     | **Hybrid life-cycle model** fills missing data via similarity regression            | Greyed number & tooltip “est. from closest match” |
| Recommendation | **Counterfactual LLM** suggests lower-impact substitutes (same category / function) | Inline “Switch & save 40 % CO₂” pill              |
| Behaviour      | **Gamified carbon budget** model predicts user motivation & nudges                  | Progress ring toward monthly goal                 |

### Key screens & states

1. **Live Cart View**

   * Item rows with CO₂ badge, uncertainty icon.
   * Toggle column to show price vs emissions.
2. **Alternative Suggestions Modal**

   * Carousel of lower-impact items; comparison metric chips (price, rating, CO₂).
3. **Checkout Summary**

   * Total footprint, shipping option selector with delta.
   * Checkbox “Auto-offset for ₹xx”.
4. **Impact Dashboard**

   * Monthly bar chart of CO₂ vs personal target.
   * Shareable badge / social post button.
5. **Post-Purchase Email Template** (non-UI artifact)

   * Receipt + climate comparison “You saved X km of driving”.

## 6. “ContextVocab” — Instant Vocabulary Companion

### Problem statement

* **Who**: Intermediate language learners reading web articles or e-books on a phone.
* **Pain**: Jumping to a dictionary app breaks flow; plain definitions lack context; spaced-repetition decks are manual.
* **Scope**: Single-word lookup (phrases optional); works offline after model download.

### AI-enabled features

| Function      | Capability                                                           | UX Detail                               |
| ------------- | -------------------------------------------------------------------- | --------------------------------------- |
| Capture       | **On-device OCR & tap-to-select**                                    | Edge glow around recognised word        |
| Semantics     | **Context-aware LLM** gives definition, nuance, register             | Card answers “Why this word here?”      |
| Pronunciation | **Neural TTS** with speed slider                                     | Waveform button turns to pause icon     |
| Memory        | **Adaptive spaced-repetition scheduler** using learner error history | Progress ring on word-bank list         |
| Grouping      | **Semantic clustering** groups synonyms/antonyms                     | Chips under Word Bank: *Travel*, *Food* |

### Key screens

1. **In-Text Pop-up Card**

   * Word, IPA, part-of-speech, 2 context examples, star icon.
   * Swipe up → expanded details.
2. **Word Bank Home**

   * List view with spaced-repetition status dots.
   * Filter chips (Mastered, Learning).
3. **Quick Quiz Screen**

   * Multiple-choice or type-the-word; confetti micro-interaction on correct.
4. **Stats Dashboard**

   * Daily streak, words mastered, weakest semantic domains.
5. **Settings**

   * Base language, daily goal, offline pack download.

## 7. “SmartReceipts” — SME Expense Scanner & Cash-Flow Forecaster

### Problem statement

* **Who**: Owners of small retail shops and service agencies who juggle expenses in shoe-boxes and ad-hoc spreadsheets.
* **Pain**: Manual entry steals evenings; missing categories wreak havoc at tax time; no forward view of cash.
* **Scope**: Receipts & invoices only; payroll, loans, and inventory valuation are excluded.

### AI-enabled features

| Stage          | Capability                                                | UX Implication                                |
| -------------- | --------------------------------------------------------- | --------------------------------------------- |
| Ingest         | **Multi-lingual OCR + layout parser**                     | Auto-highlights extracted fields; edit inline |
| Classification | **LLM with expense taxonomy** maps to GST categories      | Confidence badge & dropdown to override       |
| Forecast       | **Prophet-style time-series + scenario model**            | Interactive cash-balance graph w/ slider      |
| Alerts         | **Anomaly detector** flags outlier spend                  | Toast “Unusual expense 3× average”            |
| Export         | **Conversational report generator** (LLM) → narrative PDF | Button “Explain my cash burn”                 |

### Key screens

1. **Capture Flow**

   * Camera overlay grid; batch-scan mode with counter badge.
2. **Auto-Categorisation Table**

   * Rows: Date, Vendor, Category, GST %, Amount; pencil icon to edit.
3. **Cash-Flow Forecast Chart**

   * Line graph with shaded confidence band; slider for “What if revenue –10 %”.
4. **Alert Centre**

   * Cards: anomaly reason, link to receipt, dismiss or accept.
5. **Export & Sharing**

   * Options: CSV, PDF narrative, push to Xero/QuickBooks.

## 8. “Fit-N-Find” — AI Size & Fit Recommender for Apparel

### Problem statement

* **Who**: Fashion e-commerce shoppers frustrated by returns; retailers bleeding logistics cost and reverse-logistics emissions.
* **Pain**: Size charts differ by brand; body measurements are hard to take; style preferences matter.
* **Scope**: Tops, bottoms, dresses; excludes shoes & accessories.

### AI-enabled features

| Layer       | Capability                                                                  | UX Hook                                                |
| ----------- | --------------------------------------------------------------------------- | ------------------------------------------------------ |
| Capture     | **Privacy-preserving body-scan via phone camera** (depth-from-video + pose) | Progress ring “95 % complete”; local processing banner |
| Match       | **Similarity model** joins scan + past keep/return history                  | Confidence label “High fit certainty”                  |
| Brand adapt | **LLM maps brand size tables to universal body dims**                       | Tooltip “Brand runs small; recommending M”             |
| Feedback    | **Reinforcement loop** asks post-delivery fit feedback                      | Emoji scale; improves model                            |
| Style       | **Vision-language model** for style similarity (optional stretch)           | “You may also like” gallery                            |

### Key screens

1. **Body-Scan On-boarding**

   * Privacy explainer, light guidance video, start button.
2. **Live Scan**

   * Silhouette outline alignment; tip cards (“Turn slowly”).
3. **Product Page with Fit Badge**

   * “Your perfect fit: M” pill, certainty %, link “Why?”.
   * Alternate sizes dropdown greys out poor fits.
4. **Order History & Feedback**

   * Timeline of purchases with keep/return toggle; quick thumbs-up/down fit buttons.
5. **Fit Profile Settings**

   * Re-scan option, delete data, body-shape avatar preview.
